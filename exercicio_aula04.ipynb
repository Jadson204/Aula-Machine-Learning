### Utilizando sries e dataframes, crie um quadro de dados com as seguintes informaes utilizando um dicionrio em Python:


```python
import pandas as pd
import numpy as np

dict = {
    'nome': ['Anderson', 'Beatriz', 'Bruno', 'Cristiano', 'Daniel', 'Douglas', 'Elma', 'Fabricio', 'Gabriel', 'Janaina', 'Luiz', 'Marcos', 'Nadja', 'Oswaldo', 'Patricia', 'Ricardo'],
    'idade': [24, 35, 45, 55, 29, 20, 38, 47, 23, 56, 32, 22, 36, 68, 22, 38],
    'cidade': ['Santos', 'Atibaia', 'Guarulhos', 'Atibaia', 'Santos', 'Guarulhos', 'Guarulhos', 'Atibaia', 'Guarulhos', 'Guarulhos', 'Santos', 'Atibaia', 'Guarulhos', 'Santos', 'Atibaia', 'Guarulhos'],
    'Registro': [32123, 45828, 34985, 13457, 98439, 99348, 45019, 24894, 32948, 13480, 22884, 56821, 65938, 12956, 75439, 45392],
    'Departamento': ['RH', 'PD', 'RH', 'ADM', 'ADM', 'RH', 'ADM', 'PD', 'PD', 'ADM', 'PD', 'ADM', 'PD', 'RH', 'ADM', 'PD'],
    'Formao': ['Graduao', 'Mestrado', 'Graduao', 'Tcnico', 'Graduao', 'Tcnico', 'Graduao', 'Graduao', 'Tcnico', 'Graduao', 'Graduao', 'Graduao', 'Graduao', 'Mestrado', 'Graduao', 'Mestrado']
}

df = pd.DataFrame(dict)

print(df)
```

### Utilizando o quadro de dados criado, e utilizando o modelo de agrupamento (groupby) estudado em sala, crie um grupamento utilizando:
-  as cidades dos funcionrios;

```python
grupo_cidade = df.groupby('cidade')

df['dif_idade_cidade'] = grupo_cidade['idade'].transform(lambda x: x - x.mean())

print(df[['nome', 'cidade', 'idade', 'dif_idade_cidade']])
```

- a formao dos funcionrios;

```python
grupo_formacao = df.groupby('Formao')

df['dif_idade_formacao'] = grupo_formacao['idade'].transform(lambda x: x - x.mean())

print(df[['nome', 'Formao', 'idade', 'dif_idade_formacao']])
```

- o departamento dos funcionrios;

```python
grupo_departamento = df.groupby('Departamento')

df['dif_idade_departamento'] = grupo_departamento['idade'].transform(lambda x: x - x.mean())

print(df[['nome', 'Departamento', 'idade', 'dif_idade_departamento']])
```

- a cidade e o departamento dos funcionrios;

```python
grupo_cidade_departamento = df.groupby(['cidade', 'Departamento'])

df['dif_idade_cidade_departamento'] = grupo_cidade_departamento['idade'].transform(lambda x: x - x.mean())

print(df[['nome', 'cidade', 'Departamento', 'idade', 'dif_idade_cidade_departamento']])
```

- o departamento e a formao dos funcionrios;

```python
grupo_departamento_formacao = df.groupby(['Departamento', 'Formao'])

df['dif_idade_departamento_formacao'] = grupo_departamento_formacao['idade'].transform(lambda x: x - x.mean())

print(df[['nome', 'Departamento', 'Formao', 'idade', 'dif_idade_departamento_formacao']])
```
